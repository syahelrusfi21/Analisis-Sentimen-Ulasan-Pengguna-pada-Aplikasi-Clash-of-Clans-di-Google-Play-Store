```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort

# Mengimpor pustaka pandas untuk mengolah data ulasan dalam format DataFrame
import pandas as pd
```

# Scapping Dataset

Dalam proyek ini, kita akan menganalisis sentimen dari ulasan dan pendapat pengguna terkait aplikasi game "Clash of Clans" di Google Play Store. Tujuan dari analisis ini adalah memahami bagaimana pengguna merasakan serta mengungkapkan opini mereka terhadap aplikasi tersebut.

Dengan memanfaatkan berbagai teknik pemrosesan teks serta algoritma machine learning, kita akan menentukan apakah ulasan pengguna mengenai "Clash of Clans" bersifat positif, negatif, atau netral. Hasil dari analisis ini dapat memberikan wawasan berharga bagi pengembang dalam meningkatkan pengalaman pengguna. Mari kita eksplorasi ulasan pengguna dan mendalami analisis sentimen seputar aplikasi game "Clash of Clans" dengan lebih mendalam.

```python
# Scraping review
scrapreview, _ = reviews(
    'com.supercell.clashofclans', # ID aplikasi
    lang='id',                    # Bahasa ulasan (default: 'en')
    country='id',                 # Negara (default: 'us')
    sort=Sort.MOST_RELEVANT,      # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=12000                   # Jumlah ulasan yang ingin diambil
)

# Convert ke DataFrame
df = pd.DataFrame(scrapreview)

# Simpan ke CSV
df.to_csv('review_clashofclans.csv', index=False, encoding='utf-8-sig')
```
